```python
import csv
import json
import pandas as pd


with open('Telco-Customer-Churn.json', encoding='utf-8') as inputfile:
    data = json.loads(inputfile.read())

df = pd.json_normalize(data)



```

```python
df.head()
```

**Arrumando os Nomes das Colunas**


```python
df.rename(columns={"customerID" : "clienteId",
                   "Churn":"cancelouPlano",
                   "customer.gender":"genero",
                   "customer.SeniorCitizen":"cidadaoIdoso",
                   "customer.Partner":"possuiParceiro",
                   "customer.Dependents":"possuiDependentes",
                   "customer.tenure":"messesContrato",
                   "phone.PhoneService":"servicoTelefone",
                   "phone.MultipleLines":"multiplasLinhas",
                   "internet.OnlineSecurity":"segurancaOnline",
                   "internet.InternetService":"servicoInternet",
                   "internet.OnlineBackup":"backupOnline",
                   "internet.DeviceProtection":"protecaoDispositivo",
                   "internet.TechSupport":"suporteTecnico",
                   "internet.StreamingTV":"streamingTv",
                   "internet.StreamingMovies":"streamingFilmes",
                   "account.Contract":"tipoContrato",
                   "account.PaperlessBilling":"contaOnline",
                   "account.PaymentMethod":"metodoPagamento",
                   "account.Charges.Monthly":"totalGastoMes",
                   "account.Charges.Total":"totalGasto"}, inplace=True)
```

```python
df.head()
```

**Busca de Valores Nulos**

```python
for col in df.columns:
  vazio = df[col].isnull().sum()
  
  print(f'A coluna {col} possui {vazio} registros vazios')
```

**Capturando informações sobre o Dataset**

```python
df.info()
```

```python
for col in df.columns:
  for lin, posicao in zip(df[col], range(df[col].shape[0])):
    if lin == " ":
      df.loc[posicao,col] = 0
      print('Alterado')
    if lin == '':
      df.loc[posicao,col] = 'Sem Informação'
      print('Alterado')
```

**Alterando Tipos das Colunas para Numerico e float**

```python
from numpy import float64
#df['totalGasto'] = df['totalGasto'].astype('float64')
df['totalGasto'] = df['totalGasto'].map(float64)
df['totalGasto'].unique()
```

```python
df['cidadaoIdoso'] = df['cidadaoIdoso'].replace([0],'Não')
df['cidadaoIdoso'] = df['cidadaoIdoso'].replace([1],'Sim')

df['cidadaoIdoso'].unique()
```

**TIPOS DE DADOS**

```python
df.info()
```

```python
for col in df.columns:
  print(f' A {col} tem esses registro:{df[col].unique()}')
```

**Alterações de Valores**

```python
for col in df.columns:
  for lin, posicao in zip(df[col], range(df[col].shape[0])):
    if lin == 'No':
      df.loc[posicao,col] = 'Não'
    if lin == 'Yes':
      df.loc[posicao,col] = 'Sim'
    if lin == 'Female':
      df.loc[posicao,col] = 'Feminino'
    if lin == 'Male':
      df.loc[posicao,col] = 'Masculino'
    if lin == 'No phone service':
      df.loc[posicao,col] = 'Sem serviço de telefone'
    if lin == 'Fiber optic':
      df.loc[posicao,col] = 'Fibra ótica'
    if lin == 'No internet service':
      df.loc[posicao,col] = 'Sem serviço de internet'
    if lin == 'One year':
      df.loc[posicao,col] = 'Um ano'
    if lin == 'Month-to-month':
      df.loc[posicao,col] = 'Mensal'
    if lin == 'Two year':
      df.loc[posicao,col] = 'Dois anos'
    if lin == 'Mailed check':
      df.loc[posicao,col] = 'Cheque enviado'
    if lin == 'Electronic check':
      df.loc[posicao,col] = 'Cheque eletronico'
    if lin == 'Credit card (automatic)':
      df.loc[posicao,col] = 'Cartão de crédito (automático)'
    if lin == 'Bank transfer (automatic)':
      df.loc[posicao,col] = 'Transferência bancária (automático)'

```
